### Data cleaning

#### Year_of_Release

```python
df['Year_of_Release'].replace('9/17/2013', '2013', inplace=True)
```

```python
df['Year_of_Release'].fillna(0, inplace=True)
```

```python
df[df['Year_of_Release'].isnull()]
```

#### Genre

```python
df['Genre'].fillna('Not_informed', inplace=True)
df['Genre'].unique()
```

#### Publisher

```python
df['Publisher'] = df['Publisher'].str.title()
```

```python
df['Publisher'].replace('Ñintendo', 'Nintendo', inplace=True)
```

```python
df['Publisher'].unique()
```

```python
df['Publisher'].value_counts().plot(kind="bar")
```

#### Global_Sales

```python
pd.set_option('display.max_rows', None)
```

```python
df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].head(10)
```

```python
df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].fillna(0, inplace=True)
```

```python
df['NA_Sales'] = df['NA_Sales'].apply(lambda x: float(str(x).replace(',', '.')) if isinstance(x, str) else x).round(2)
df['EU_Sales'] = df['EU_Sales'].apply(lambda x: float(str(x).replace(',', '.')) if isinstance(x, str) else x).round(2)
df['JP_Sales'] = df['JP_Sales'].apply(lambda x: float(str(x).replace(',', '.')) if isinstance(x, str) else x).round(2)
df['Other_Sales'] = df['Other_Sales'].apply(lambda x: float(str(x).replace(',', '.')) if isinstance(x, str) else x).round(2)
df['Global_Sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
```

### Data transformation

```python
df1['PlatformType'] = df1['Platform'].apply(lambda x: 'Home Console' if x in ['Wii', 'NES', 'SNES', 'N64', 'X3600000', 'X360', 'XB', 'XOne','PS3', 'PS2', 'PS', 'PS4'] else
                                            'Handheld Console' if x in ['GB', 'DS', 'GBA', '3DS', 'PSP'] else
                                            'Retro Console' if x == '2600' else
                                            'PC' if x == 'PC' else
                                            None
                                            )
```

```python
df1['GenreType'] = df1['Genre'].apply(lambda x: 'Action and Combat' if x in ['Platform', 'Action', 'Shooter', 'Fighting'] else
                                      'Sports and Racing' if x in ['Sports', 'Racing'] else
                                      'Adventure and Exploration' if x in ['Adventure', 'Role-Playing'] else
                                      'Strategy and Simulation' if x in ['Puzzle', 'Simulation'] else
                                      'Miscellaneous' if x == 'Misc' else
                                      None)
```

```python
df1['PublisherType'] = df1['Publisher'].apply(lambda x: 'First-Party Publishers' if x in ['Nintendo', 'Microsoft Game Studios', 'Sony Computer Entertainment'] else
                                              'Third-Party Publishers' if x in ['Electronic Arts', 'Take-Two Interactive', 'Activision', 'Ubisoft'] else
                                              'RPG Publishers' if x in ['Bethesda Softworks', 'SquareSoft'] else
                                              'Classic Publishers' if x in ['Sega', 'Atari'] else
                                              None)
```

```python
df1['PlatformType'].value_counts().plot(kind="bar", legend=True, xlabel="", figsize=(8, 6), rot=0, title="Platform Type")
```

```python
df1['GenreType'].value_counts().plot(kind="bar", legend=True, xlabel="", figsize=(12, 6), rot=0, title="Genre Type")
```

```python
df1['PublisherType'].value_counts().plot(kind="bar", legend=True, xlabel="", figsize=(10, 6), rot=0, title="Publisher Type")
```
